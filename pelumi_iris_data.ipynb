```python
#plotting a line plot for the dataset
plt.plot(df.sepal_length, df.sepal_width, label="Sepal plot", marker='s')
plt.plot(df.petal_length, df.petal_width, label="Petal plot",marker='s')
plt.title("Line Graph Showing the graph of length plot against width ")
plt.xlabel(" Length")
plt.ylabel(" Width")
plt.legend()
plt.show()
```
